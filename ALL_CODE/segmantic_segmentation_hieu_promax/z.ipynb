```python
import json
from cryptography.fernet import Fernet

# Tạo khóa bí mật
key = Fernet.generate_key()
cipher_suite = Fernet(key)

# Gán giá trị cho biến y và x
y = "Khoảng thời gian"
x = 1

# Mã hóa dữ liệu
encrypted_y = cipher_suite.encrypt(y.encode())
encrypted_x = cipher_suite.encrypt(str(x).encode())

# Tạo dictionary chứa dữ liệu đã mã hóa
data = {
    "encrypted_y": encrypted_y.decode(),
    "encrypted_x": encrypted_x.decode()
}

# Ghi dữ liệu vào file JSON
with open("encrypted_data.json", "w") as json_file:
    json.dump(data, json_file)
```

```python
import json
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from base64 import b64encode, b64decode

# Dữ liệu ban đầu
data = {
    "sensitive_info": "This is confidential data!",
    "a": "b",
    "c":"d"
}
secret_key = "MySecretKey123456"  # Khóa bí mật

# Mã hóa dữ liệu
cipher = AES.new(secret_key.encode(), AES.MODE_CBC)
cipher_text = cipher.encrypt(json.dumps(data).encode())

# Lưu dữ liệu vào file JSON
with open("encrypted_data2.json", "wb") as json_file:
    json_file.write(b64encode(cipher.iv + cipher_text))

# Đọc dữ liệu từ file JSON
with open("encrypted_data2.json", "rb") as json_file:
    encrypted_data = b64decode(json_file.read())

# Giải mã dữ liệu
iv = encrypted_data[:AES.block_size]
cipher_text = encrypted_data[AES.block_size:]
decipher = AES.new(secret_key.encode(), AES.MODE_CBC, iv)
decoded_data = json.loads(decipher.decrypt(cipher_text).decode())

# In kết quả giải mã
print(decoded_data)
```
