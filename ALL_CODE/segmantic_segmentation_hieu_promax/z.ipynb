```python
import json
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from base64 import b64encode, b64decode

# D liu ban u
data = {
    "sensitive_info": "This is confidential data!",
    "a": "b",
    "c":"d"
}
secret_key = "MySecretKey123456"  # Kha b mt

# M ha d liu
cipher = AES.new(secret_key.encode(), AES.MODE_CBC)
cipher_text = cipher.encrypt(json.dumps(data).encode())

# Lu d liu vo file JSON
with open("encrypted_data2.json", "wb") as json_file:
    json_file.write(b64encode(cipher.iv + cipher_text))

# c d liu t file JSON
with open("encrypted_data2.json", "rb") as json_file:
    encrypted_data = b64decode(json_file.read())

# Gii m d liu
iv = encrypted_data[:AES.block_size]
cipher_text = encrypted_data[AES.block_size:]
decipher = AES.new(secret_key.encode(), AES.MODE_CBC, iv)
decoded_data = json.loads(decipher.decrypt(cipher_text).decode())

# In kt qu gii m
print(decoded_data)
```
